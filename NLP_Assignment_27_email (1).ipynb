```python
data=pd.read_csv('SMSSpamCollection', sep="\t", names=['label','message'])
data.info()
```

```python
import nltk
nltk.download('stopwords')
```

```python
data.head()
```

```python
import nltk
from nltk.corpus import stopwords
import string

def textPreprocessor(featureRecords):
    #a.Remove Punctuation
    removePunctuation = [char for char in featureRecords if char not in string.punctuation]
    sentences = ''.join(removePunctuation)

    #b.Convert Sentectes to Words
    words = sentences.split(" ")

    #c.Normalize
    wordNormalized = [word.lower() for word in words]

    #d. Remove Stopwords
    finalWords = [word for word in wordNormalized if word not in stopwords.words("english")]

    return finalWords
```

```python
textPreprocessor("Welcome to SIT! You are learning NLP")
```

```python
features = data.iloc[:,1].values
label = data.iloc[:,0].values
```

```python
print(features)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
wordVector = CountVectorizer(analyzer=textPreprocessor)
finalWordVocab = wordVector.fit(features)
```

```python
finalWordVocab.vocabulary_
```

```python
bagOfWords = finalWordVocab.transform(features)
```

```python
bagOfWords
```

```python
from sklearn.feature_extraction.text import TfidfTransformer
tfIdfObject = TfidfTransformer().fit(bagOfWords)
finalFeature = tfIdfObject.transform(bagOfWords)
```

```python
print(finalFeature)
```

```python
from sklearn.model_selection import train_test_split
X_Train,X_Test,Y_Train,Y_Test=train_test_split(finalFeature,label,test_size=0.18,random_state=6)
```

```python
Y_Train
```

```python
from sklearn.linear_model import LogisticRegression
ml=LogisticRegression()
ml.fit(X_Train,Y_Train)
```

```python
print("accuracy: ",ml.score(X_Test,Y_Test)*100)
```
